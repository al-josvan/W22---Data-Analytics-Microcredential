```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/property.csv')
```

```python
data
```

```python
#Filling in missing property ID
new_PID = data['PID'].fillna(100005000.0)
```

```python
#Removing previous PID label
data.drop(labels = 'PID', axis = 1, inplace= True)
```

```python
#Adding new PID with missing value filled in
data['PID'] = new_PID
```

```python
#Seeting PID as index 
data.set_index('PID', inplace = True)
```
